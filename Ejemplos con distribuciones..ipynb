Para las distribuciones chi-cuadrado, únicamente se necesita el número de grados de libertad. 

```r
# Genera 100 elementos de una distribución Chi-cuadrado
set.seed(123)
datos <- rchisq(100, df = 3)

# Crear el histograma
suppressWarnings(hist(datos, breaks = 10, freq = FALSE, main = "Histograma", xlab = "Valores", ylab = "Densidad"))

# Superponer la curva teórica
x_vals <- seq(0, max(datos) + 10, length.out = 100)  # Valores en el rango deseado
curva_teorica <- curve(dchisq(x, df = 3), from = min(x_vals), to = max(x_vals), add = TRUE, col = "cyan3", lwd = 4)

# Agregar leyenda
legend("topright", legend = "Curva teórica", col = "cyan3", lwd = 4)
```
